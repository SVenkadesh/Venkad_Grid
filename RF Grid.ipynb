```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
Param_grid = {'criterion':['squared_error', 'absolute_error', 'friedman_mse', 'poisson'], 'max_features':['sqrt', 'log2', 'None']}
RF_grid = GridSearchCV(RandomForestRegressor(),Param_grid,refit=True,verbose=3,n_jobs=-1)
RF_grid.fit(indep,dep)
```

```python
prnt=RF_grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(prnt)
table
```

```python
print("The R_score value for best parameter {}:".format(RF_grid.best_params_))
```
